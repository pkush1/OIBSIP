```python
import pandas as pd
df2 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
print(df2)
```

```python
df2 = df2.rename(columns={df2.columns[0]:'Region',df2.columns[3]:'Unemployment_rate',df2.columns[4]:'Employed', df2.columns[5]:'labour_participation_rate', df2.columns [6]:'area'})
df2.head()
```

```python
import seaborn as sns
heat_maps = df[['Unemployment rate','Employed','labour_participation_rate']]
heat_maps = heat_maps.corr()

plt.figure(figsize=(11,6))
sns.set_context('notebook',font_scale=1)
sns.heatmap(heat_maps, annot=True,cmap='winter');
```

```python
import seaborn as sns
df2.columns=["Region", "Date", "Frequency", "Unemployment rate", "Employed", "labour_participation_rate", "area", "longitude", "latitude"]
plt.figure(figsize=(12,16))
plt.title("Unemployment rate")
sns.histplot(x="Unemployment rate", hue="Region", data=df)
plt.show()
```

```python
region = df2.groupby(["Region"])[['Unemployment rate', "Employed", "labour_participation_rate"]].mean()
region = pd.DataFrame(region).reset_index()

fig= px.bar(region, x="Region", y="Employed", color="Region", title="Average Employed Rate by Region")
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
region = df2.groupby(["Region"])[['Unemployment rate', "Employed", "labour_participation_rate"]].mean()
region = pd.DataFrame(region).reset_index()
fig = px.bar(region, x="Region", y="Unemployment rate", color="Region", title="Average Employed Rate by Region")
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
unemployment = df2[["Region", "area", "Unemployment rate"]]
fig = px.sunburst(unemployment, path=['area', 'Region'], values='Unemployment rate',title='Unempleyment rate in every state and Region', height=700)
fig.show()
```
